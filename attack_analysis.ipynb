```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')
```

```python
min_played = [2600, 2701, 1528, 1729]
player = ['Benzema', 'Vinicius', 'Rodrygo','Asensio']
color_list = ['red','blue','green','pink']
plt.bar(player , min_played , label = 'Minutes Played', color = color_list)
plt.legend()
plt.title('Minutes Played')
plt.xlabel('Players')
plt.ylabel('Minutes Played')
plt.show()
```
